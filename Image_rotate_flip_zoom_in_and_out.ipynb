```python
import matplotlib.image as image
from PIL import Image
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
img=image.imread('download.jpg')
print('The Shape of the image is:',img.shape)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #converting image into grayscale image
arr= np.array(gray)

#dimensions of gray scal image array
c=len(arr[0])
we=len(arr)
#print(c)
#print(we)

arr2 = [[] for i in range(c)] #declaring arr2 for stroing rotated image matrix
#dimensions of this matrix not needed just for checking row size to avoid index out of range error
#ee=len(arr2[0])
#ff=len(arr2)
#print(ee)
#print(ff)

# Transposing matrix of given image
for i in range(c): # loop for row iteration
   for j in range(we):#loop for column iteration
      val=arr[j][i]
      arr2[i].append(val)


print(arr.shape)
print("Original gray Image")
cv2_imshow(arr)

#converting arr2 (rotated image matrix) as numpy 
arr2=np.array(arr2)
cv2.imwrite("rotated.png", arr2) #saving rotated image
#print("Matrix of rotated image")
#print(arr2)
print("Rotated image")
cv2_imshow(arr2)

#---------------Flipping  image by reversing the matrix-----------
fliparr=[[] for i in range(we)] 
arr4=[[] for i in range(we)]

for i in range(len(arr)):
  for j in range(c-1,-1,-1):#traversing column in reverse
    val=arr[i][j]   #stores the value of last column
    fliparr[i].append(val) #adds that value at end of list i.e, if list is empty that the last value or arr matrix coumn comes at first column in fliparr


fliparr=np.array(fliparr)
print("flipped image")
cv2_imshow(fliparr)
   

#img4 = Image.open('rotated.png')
#img4= img4.resize((160, 240), Image.ANTIALIAS)
#img4.show()
#zooming image in
basewidth = 600
imgzoom = Image.open('download.jpg')
wpercent = (basewidth/float(imgzoom.size[0]))
hsize = int((float(imgzoom.size[1])*float(wpercent)))
imgzoomin = imgzoom.resize((basewidth,hsize), Image.ANTIALIAS)
imgzoomin.save('zoomin.jpg')
imgzi=image.imread('zoomin.jpg')

print("Original")
cv2_imshow(img)
print("Zoomed in")
cv2_imshow(imgzi)

#zooming image out

half = 0.5
out = imgzoom.resize( [int(half * s) for s in imgzoom.size] )
out.save('zoomout.jpg')
imgzo=image.imread('zoomout.jpg')
print("Zoomed out")
cv2_imshow(imgzo)


```
